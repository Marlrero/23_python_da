```python
language = col_data['LanguageHaveWorkedWith']
language
```

```python
language = language.str.split(';')
language
```

```python
exploded_language = language.explode()
exploded_language
```

```python
size_by_lan = exploded_language.groupby(exploded_language).size()
size_by_lan
```

```python
exploded_language.groupby(exploded_language).size().sort_values(ascending=False)
```

```python
size_by_lan.nlargest(10).plot.pie(figsize=(10, 10))
```

```python
size_by_lan.nlargest(10).plot.pie(figsize=(10, 10), autopct='%1.0f%%')
```

```python
lan_for_25_34 = col_data[col_data.Age == '25-34 years old']
lan_for_25_34
```

```python
lan_for_25_34 = lan_for_25_34['LanguageHaveWorkedWith'].str.split(';').explode()
lan_for_25_34
```

```python
lan_for_25_34.groupby(lan_for_25_34).size().nlargest(10) \
  .plot.pie(figsize=(10, 10), autopct='%1.0f%%')
```
